#### Primer Conjunto de Datos: Global Temperature

Explorando los datos

```python
import pandas as pd

# Cargar el conjunto de datos de temperatura global
df_temp = pd.read_csv('./data/Global_Temp.csv')

# Shape
df_temp.shape

# Mostrar las primeras filas para explorar la estructura
df_temp.head()

```

Mensaje a transmitir:

La intención es mostrar la tendencia de cambio en la temperatura global a lo largo de los años, destacando los aumentos o descensos significativos y analizando posibles patrones de calentamiento global.

Selección del Tipo de Gráfica:

Para ilustrar la tendencia a lo largo del tiempo, un gráfico de líneas es apropiado, ya que permite ver la variación en la temperatura de un año a otro y observar una tendencia general.

Gráfica inicial

```python
import matplotlib.pyplot as plt

# Convertir la columna de "Year" a índice para facilitar la visualización
df_temp.set_index('Year', inplace=True)

# Crear el gráfico de líneas usando la columna "J-D" (promedio de temperaturas anuales)
plt.figure(figsize=(12, 6))
plt.plot(df_temp.index, df_temp['J-D'], color='b', linewidth=1.5, label='Promedio Anual')

# Etiquetas y título
plt.title('Tendencia de la Temperatura Global (1880-2023)', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Anomalía de Temperatura (°C)', fontsize=12)
plt.legend()

# Mostrar la gráfica
plt.show()
```

Optimizando carga cognitiva

```python
import numpy as np

# Crear el gráfico de líneas con optimización en el eje Y
plt.figure(figsize=(12, 6))
plt.plot(df_temp.index, df_temp['J-D'], color='b', linewidth=1.5, label='Promedio Anual')

# Etiquetas y título
plt.title('Tendencia de la Temperatura Global (1880-2023)', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Anomalía de Temperatura (°C)', fontsize=12)
plt.legend()

# Ajustar el eje Y para que tenga menos marcas y sea más legible
# plt ytciks to go en aumentos de 10
plt.yticks(np.arange(-0.5, 100, 10))

# Mostrar la gráfica optimizada
plt.show()
```

Explicación:

Este gráfico de líneas representa la anomalía de temperatura promedio global desde 1880 hasta 2023. Elegí un gráfico de líneas para mostrar claramente cómo han variado las temperaturas a lo largo del tiempo, lo cual ayuda a visualizar tendencias a largo plazo y resaltar períodos de calentamiento o enfriamiento. Al optimizar la gráfica con etiquetas claras en el eje y, reduciendo la cantidad de steps en la escala, la carga cognitiva se reduce, permitiendo que el espectador se concentre en la tendencia general en lugar de detalles innecesarios.

#### Segundo Conjunto de Datos: Historical CO2 Emissions

Explorando los datos

```python
# Cargar el conjunto de datos de emisiones históricas de CO₂
df_emissions = pd.read_csv('./data/historical_emissions.csv')

# Mostrar las primeras filas para explorar la estructura
df_emissions.head()
```

Mensaje a transmitir:

Nuestra intención es mostrar cómo han cambiado las emisiones de CO₂ a nivel mundial y en algunos países clave (como China y Estados Unidos) a lo largo del tiempo. Queremos ilustrar las tendencias en emisiones y comparar países para resaltar contribuciones significativas al cambio climático.

Selección del Tipo de Gráfica:

Para ilustrar las emisiones de CO₂ a lo largo del tiempo, un gráfico de líneas es nuevamente apropiado, ya que queremos mostrar la evolución de las emisiones en función del año. Además, vamos a comparar algunos países clave, por lo que un gráfico de líneas con múltiples series es adecuado.

Gráfica Inicial:

```python
import matplotlib.pyplot as plt

# Filtrar datos para algunos países clave
countries = ['World', 'China', 'United States', 'India', 'European Union']
df_filtered = df_emissions[df_emissions['Country'].isin(countries)]

# Transponer los datos para tener los años como filas
df_filtered = df_filtered.set_index('Country').iloc[:, 5:].transpose()
df_filtered.index = df_filtered.index.astype(int)  # Convertir años a enteros

# Crear el gráfico de líneas
plt.figure(figsize=(14, 8))

for country in countries:
    plt.plot(df_filtered.index, df_filtered[country], label=country)

# Etiquetas y título
plt.title('Tendencia de Emisiones de CO₂ por País (1990-2018)', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Emisiones de CO₂ (MtCO₂e)', fontsize=12)
plt.legend()

# Mostrar la gráfica
plt.show()
```

Optimización para Reducir la Carga Cognitiva <br>
- Colores y etiquetas claras: Utilizamos colores diferentes para cada país y añadimos una leyenda para facilitar la identificación.
- Límite de países: Seleccionamos solo algunos países clave (China, Estados Unidos, India, Unión Europea y el mundo en general) para evitar sobrecargar la gráfica.
- Rango de años: Incluimos un rango claro de años en el eje X y mantenemos el eje Y escalado adecuadamente para que las variaciones sean perceptibles sin distorsión.

Explicación:

Este gráfico de líneas muestra las emisiones de CO₂ de 1990 a 2018 para algunos países clave y el total mundial. Elegimos un gráfico de líneas con múltiples series para ilustrar cómo las emisiones han cambiado en cada país y comparar sus contribuciones al cambio climático. La optimización de la gráfica se logró limitando la cantidad de países y usando etiquetas y colores distintivos, permitiendo una comparación clara entre países sin sobrecargar la gráfica.

#### Tercer Conjunto de Datos: Tienda en linea ventas mensuales

Explorando los datos

```python
# Cargar el conjunto de datos de ventas de la tienda en línea
df_sales = pd.read_csv('./data/superstores.csv')

# Mostrar las primeras filas para explorar la estructura
df_sales.head()
```

Mensaje a transmitir:

Queremos analizar las ventas totales por categoría de producto para identificar cuáles generan mayores ingresos en la tienda. Esta información es útil para comprender el rendimiento de cada categoría y tomar decisiones estratégicas.

Selección del Tipo de Gráfica:

Para comparar las ventas entre categorías de productos, un gráfico de barras es ideal, ya que permite una comparación clara y sencilla de los totales de ventas por categoría.

Gráfica Inicial:

```python

# Calcular las ventas totales por categoría de producto
sales_by_category = df_sales.groupby('Category')['Sales'].sum().sort_values()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sales_by_category.plot(kind='barh', color=['skyblue', 'lightgreen', 'salmon'])

# Etiquetas y título
plt.title('Ventas Totales por Categoría de Producto', fontsize=16)
plt.xlabel('Ventas Totales (USD)', fontsize=12)
plt.ylabel('Categoría de Producto', fontsize=12)

# Mostrar la gráfica
plt.show()
```

Optimización para Reducir la Carga Cognitiva <br>

- Gráfico de barras horizontal: Esto hace que las categorías sean más fáciles de leer en el eje Y y permite comparar visualmente las ventas.
- Ordenar de menor a mayor: Ordenar las barras facilita identificar rápidamente qué categorías tienen menores o mayores ventas.
- Color diferente y título claro: Utilizamos un color diferente para cada categoría y un título explicativo para hacer que el gráfico sea fácil de interpretar sin distracciones.

Explicación: <br>

 Este gráfico de barras horizontal muestra las ventas totales de cada categoría de producto en la tienda en línea. Elegimos un gráfico de barras para permitir una comparación clara entre categorías, y se ordenaron las barras de menor a mayor para facilitar la identificación de las categorías con menores y mayores ventas. Este diseño minimiza la carga cognitiva y resalta el rendimiento relativo de cada categoría de producto.

Mejorando gráfica

```python

# Load the data
df = pd.read_csv('./data/ventas.csv')

# Convert the 'Fecha' column to datetime format for better handling
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%b-%y')

# Fill missing values in 'Ventas planificadas (unidades)' with the mean of the available values
df['Ventas planificadas (unidades)'].fillna(df['Ventas planificadas (unidades)'].mean(), inplace=True)

# Calculate the average of planned sales after filling missing values
avg_planned_sales = df['Ventas planificadas (unidades)'].mean()

# Plot the data
plt.figure(figsize=(12, 6))

# Plot actual sales as bars
plt.bar(df['Fecha'], df['Ventas Reales (unidades)'], color='royalblue', label='Ventas Reales')

# Plot the average of planned sales as a horizontal line
plt.axhline(avg_planned_sales, color='green', linestyle='--', linewidth=2, label='Promedio Ventas Planificadas')

# Labels and title
plt.title('Comparación de Ventas Reales vs Promedio Planificado', fontsize=16)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Unidades Vendidas (Millones)', fontsize=12)
plt.legend()

# Adjust Y-axis ticks for better readability
plt.yticks(range(0, 80000000, 10000000))

# Format the X-axis dates to show every 3 months
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%y'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator(interval=3))

# Display the optimized plot
plt.show()
```

Cambios realizados: <br>

- Escala en el Eje Y: Las unidades vendidas se dividieron entre un millón (1e6) para mostrar los valores en millones, lo que simplifica la visualización de grandes números y facilita la comparación.
- Etiqueta Actualizada del Eje Y: Se cambió la etiqueta del eje Y a "Unidades Vendidas (Millones)" para que el espectador entienda que los valores en el eje Y están expresados en millones.
- Intervalos del Eje Y Ajustados: Los intervalos del eje Y se configuraron en pasos de 10 millones para reducir el número de marcas y hacer la gráfica más legible.
- Promedio de Ventas Planificadas: La línea que representa el promedio de ventas planificadas también se ajustó a la escala de millones, manteniendo la coherencia visual con las barras de ventas reales.
- Claridad en la Comparación Temporal: Las fechas en el eje X se formatearon para mostrar solo algunos meses clave, reduciendo la sobrecarga visual y facilitando la interpretación de la evolución temporal.

Principios de la Gestalt: Cree una visualización que demuestre al menos dos

```python

data = {
    'Category': ['Tecnología', 'Tecnología', 'Tecnología', 'Muebles', 'Muebles', 'Muebles', 'Papelería', 'Papelería', 'Papelería'],
    'Sub-Category': ['Computadoras', 'Teléfonos', 'Accesorios', 'Sillas', 'Mesas', 'Estantes', 'Cuadernos', 'Plumas', 'Carpetas'],
    'Sales': [20000, 15000, 8000, 12000, 14000, 10000, 5000, 7000, 6000]
}

df_sales_fictitious = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

# Obtener las categorías únicas
categories = df_sales_fictitious['Category'].unique()

# Graficar las ventas por subcategoría, agrupadas por categoría
for category in categories:
    sub_df = df_sales_fictitious[df_sales_fictitious['Category'] == category]
    plt.bar(sub_df['Sub-Category'], sub_df['Sales'], label=category)

# Añadir etiquetas y leyenda
plt.title('Ventas por Sub-Categoría Agrupadas por Categoría (Ejemplo Ficticio)', fontsize=16)
plt.xlabel('Sub-Categoría', fontsize=12)
plt.ylabel('Ventas Totales (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Categoría")

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

Explicación de la Visualización Basada en Principios de la Gestalt <br>

- Proximidad: Las subcategorías dentro de la misma categoría están agrupadas y colocadas cerca unas de otras en el eje X. Esto permite que el espectador perciba visualmente los elementos de la misma categoría como un conjunto o grupo debido a su cercanía.

- Similitud: Todas las barras de una misma categoría tienen el mismo color, lo que ayuda a reforzar visualmente que pertenecen al mismo grupo. El espectador puede fácilmente diferenciar entre las categorías sin necesidad de leer las etiquetas de cada barra.

- Cree una gráfica que intencionalmente sea engañosa (por ejemplo, manipulando las
escalas) y explique por qué es problemática.
- Encuentre un ejemplo real de una visualización engañosa en medios de
comunicación y explique cómo podría corregirse

```python
# Datos ficticios de ventas para dos productos
productos = ['Producto A', 'Producto B']
ventas = [10000, 10200]  # Una diferencia pequeña en ventas

# Crear una gráfica engañosa limitando la escala del eje Y
plt.figure(figsize=(6, 4))
plt.bar(productos, ventas, color=['blue', 'orange'])

# Manipular la escala del eje Y para que empiece cerca de los valores mínimos
plt.ylim(9900, 10400)  # Limitar el rango para exagerar la diferencia

# Etiquetas y título
plt.title('Ventas de Productos (Gráfica Engañosa)')
plt.ylabel('Ventas (USD)')

# Mostrar la gráfica
plt.show()
```

Explicación de por qué es Problemática <br> 

- Manipulación del Eje Y: Al comenzar el eje Y en 9900 en lugar de 0, estamos exagerando la diferencia entre "Producto A" y "Producto B". Aunque la diferencia real es de solo 200 unidades, la manipulación hace que parezca mucho mayor, lo que podría llevar a interpretaciones incorrectas.

- Problema Ético: Esta técnica puede inducir a error al espectador al hacer que una diferencia pequeña parezca significativa. En un contexto comercial o de presentación de resultados, podría usarse para manipular la percepción de éxito o fracaso de un producto o servicio, lo cual es poco ético.

Ejemplo real:
![Ejemplo Real de Gráfica Engañosa](./real_example.png)

Extraído de: https://www.reddit.com/r/ottawa/comments/1gdsgf8/this_graph_by_citynews_ottawa_to_make_it_look/

Problemas detectados: <br>

- Orden Incorrecto en el Eje X:

En el gráfico, el eje X (años) no está en orden cronológico. En lugar de seguir un orden ascendente o descendente natural (por ejemplo, de 2016 a 2023 o viceversa), los años están dispuestos de manera desordenada, con un salto entre ellos que no es intuitivo.
Este orden engañoso puede llevar al espectador a pensar que los homicidios han aumentado o disminuido en una tendencia específica que en realidad no existe. Por ejemplo, la colocación del año 2016 al final con un valor alto y luego de 2020 (con el valor más bajo) da la impresión visual de un aumento dramático, aunque no sea el caso real si el orden estuviera correctamente establecido.

- Manipulación de la Percepción de Tendencia:

Al no mostrar los años en secuencia correcta, el gráfico manipula la percepción de la tendencia en el número de homicidios. Si el gráfico estuviera en orden cronológico, sería más fácil para el espectador identificar la verdadera tendencia (por ejemplo, si los homicidios están aumentando, disminuyendo o manteniéndose estables con el tiempo).

Este diseño puede llevar a conclusiones incorrectas, especialmente si el espectador asume que los datos siguen un orden cronológico cuando en realidad no lo hacen.

e) Atributos preatentivos: Cree una visualización que utilice efectivamente al menos tres
atributos preatentivos diferentes

Color: Utilizar diferentes colores para destacar categorías o valores específicos.
Forma: Usar formas distintas para representar diferentes tipos de datos.
Tamaño: Variar el tamaño de los elementos para indicar importancia o magnitud.

```python

# Crear un conjunto de datos ficticio de ventas por región y desempeño
data = {
    'Region': ['Norte', 'Sur', 'Este', 'Oeste'],
    'Sales': [30000, 15000, 20000, 35000],
    'Performance': ['Alto', 'Bajo', 'Medio', 'Alto'],
    'Satisfaction': [8.5, 6.0, 7.2, 9.0]  # Escala de 1 a 10
}


df = pd.DataFrame(data)

# Definir colores según el desempeño
color_map = {'Bajo': 'red', 'Medio': 'yellow', 'Alto': 'green'}
colors = df['Performance'].map(color_map)

# Definir tamaños de las burbujas según las ventas
sizes = df['Sales'] / 100  # Ajustar escala para el tamaño de los círculos

# Crear la visualización
plt.figure(figsize=(10, 6))

# Graficar las burbujas
plt.scatter(df['Region'], df['Satisfaction'], s=sizes, c=colors, alpha=0.6, edgecolors="w", linewidth=2)

# Añadir etiquetas de texto para las regiones
for i, row in df.iterrows():
    plt.text(row['Region'], row['Satisfaction'] + 0.2, row['Region'], ha='center', fontsize=10)

# Añadir etiquetas y título
plt.title('Ventas y Satisfacción por Región', fontsize=16)
plt.xlabel('Región', fontsize=12)
plt.ylabel('Satisfacción del Cliente', fontsize=12)

# Crear leyenda para el desempeño
import matplotlib.patches as mpatches
alto_patch = mpatches.Patch(color='green', label='Alto Desempeño')
medio_patch = mpatches.Patch(color='yellow', label='Desempeño Medio')
bajo_patch = mpatches.Patch(color='red', label='Bajo Desempeño')
plt.legend(handles=[alto_patch, medio_patch, bajo_patch], title="Desempeño", loc='upper left')

# Ajustar los límites del eje Y
plt.ylim(5, 10)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

Explicación de la Visualización y Atributos Preatentivos <br>

1. Color (Atributo Preatentivo):

    - Cada burbuja tiene un color que representa el nivel de desempeño:
Verde: Alto desempeño.
Amarillo: Desempeño medio.
Rojo: Bajo desempeño.
El uso del color permite que el espectador identifique rápidamente el desempeño de cada región.


2. Tamaño (Atributo Preatentivo):

    - El tamaño de cada burbuja está proporcionalmente relacionado con las ventas en esa región.
    Regiones con ventas más altas tienen burbujas más grandes.
    Esto ayuda a captar rápidamente cuáles son las regiones con mayores ventas.


3. Posición en el Eje Y (Atributo Preatentivo de Posición Vertical):

    - La posición vertical de las burbujas representa la satisfacción del cliente en cada región.
    Permite comparar fácilmente los niveles de satisfacción entre regiones.

Conclusión <br>
Esta visualización utiliza efectivamente tres atributos preatentivos:

- Color: Indica el nivel de desempeño.
- Tamaño: Representa el volumen de ventas.
- Posición en el eje Y: Muestra la satisfacción del cliente.
